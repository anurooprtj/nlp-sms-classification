## Visualizing Data

```python
import matplotlib.patheffects as PathEffects

import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('muted')
sns.set_context("notebook", font_scale=1.5,
                rc={"lines.linewidth": 2.5})

def sms_scatter(x, colors):
    # choose a color palette with seaborn.
    num_classes = len(np.unique(colors))
    palette = np.array(sns.color_palette("hls", num_classes))

    # create a scatter plot.
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40, c=palette[colors.astype(np.int)])
    plt.xlim(-25, 25)
    plt.ylim(-25, 25)
    ax.axis('off')
    ax.axis('tight')

    # add the labels for each digit corresponding to the label
    txts = []

    for i in range(num_classes):

        # Position of each label at median of data points.

        xtext, ytext = np.median(x[colors == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=24)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)

    return f, ax, sc, txts
```

###  1. PCA - dtm

```python
X_train_dtm_array = X_train_dtm.toarray()
```

```python
time_start = time.time()

sms_pca_dtm = PCA(n_components = 2)

X_train_dtm_pca = sms_pca_dtm.fit_transform(X_train_dtm_array) #PCA accepts np array as input.
print ('PCA done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
sms_scatter(X_train_dtm_pca, y_train)
```

### 2. t-SNE - dtm

```python
time_start = time.time()

sms_dtm_tsne = TSNE(random_state=42).fit_transform(X_train_dtm_array)

print ('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
print("type(sms_dtm_tsne): ", type(sms_dtm_tsne))
print("sms_dtm_tsne.shape: ", sms_dtm_tsne.shape)
```

```python
sms_scatter(sms_dtm_tsne, y_train)
```

### PCA - tfid 

```python
X_train_array_tfidf = X_train_tfidf.toarray()
```

```python
time_start = time.time()

sms_tfid_pca = PCA(n_components = 2)

X_train_tfidf_pca = sms_tfid_pca.fit_transform(X_train_array_tfidf) #PCA accepts np array as input.
print ('PCA tfid features done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
sms_scatter(X_train_tfidf_pca, y_train)
```

### t-SNE - tfidf 

```python
time_start = time.time()

sms_tfidf_tsne = TSNE(random_state=42).fit_transform(X_train_array_tfidf)

print ('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

```python
sms_scatter(sms_tfidf_tsne, y_train)
```
